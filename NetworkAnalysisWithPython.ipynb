# Network Analysis with Python

## Imports

```python
import numpy as np
from numpy import genfromtxt
import sys
import matplotlib.pyplot as plt
import networkx as nx
```

## Load the data

```python
adjacency_matrix = genfromtxt('adjacency_matrix.csv', delimiter=',')
```

## Create a directed graph from the adjacency matrix

```python
G=nx.from_numpy_matrix(adjacency_matrix)
G.edges(data=True)
```

## Plot the graph

```python
nx.draw(G)
plt.show()
```

## Calculate network diameter and average closeness

```python
print("Diameter: ", nx.diameter(G))
print("AVG Closeness: ", sum(nx.closeness_centrality(G).values())/25)
```

## Calculate average network betweeness

```python
print("AVG betweeness: ", sum(nx.betweenness_centrality(G))/25)
```

## Calculate network density and average degree

```python
print("Density: ", nx.density(G))
print("Average degree: ", sum(nx.degree_centrality(G).values())/25)
```

## Calculate network reciprocity and average transitivity

```python
print("Reciprocity: ", nx.reciprocity(G))
print("Transitivity: ", nx.transitivity(G))
```

## What is the average distance between two nodes?

```python
print("Average distance between two nodes: ", nx.average_shortest_path_length(G))
```
